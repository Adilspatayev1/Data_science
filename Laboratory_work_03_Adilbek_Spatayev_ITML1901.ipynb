<b>Task 2.1</b> Plotting bar charts
<br/>
On the example below, the bar chart for Seasons was plotted the documentation [here](https://matplotlib.org/3.1.1/api/_as_gen/matplotlib.pyplot.bar.html)

```python
plt.bar(data.Season.unique(), data.groupby('Season')['Sales'].sum(), width = 0.9, 
        color='#64fd00', ec='#414142')
plt.xlabel('Seasons')
plt.ylabel('Sales amount')
plt.title('Sales over seasons')
plt.show()
```

<b>Task 2.3</b> Plotting bar charts
<br/> Plot two bar charts visualizing <b>Any month days + Sales </b> and <b> June days + Sales</b>
Use subplot, both graphs should be on the same figue
<br/>
And write the summary below

```python
data['Date'] = pd.to_datetime(data['Date'])
data
```

```python
data['month'] = data.Date.dt.month_name()
data
```

```python
# Your code goes here
plt.figure(figsize=(15, 3))

plt.subplot(1, 3, 1)
data.groupby([data['Date'].dt.month_name()], sort=False)['Sales'].sum().plot(kind='bar')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.title('Sales over Month')
plt.subplot(1, 3, 2)
data[data.month == 'July'].groupby('Day')['Sales'].sum().plot(kind='bar')
plt.xlabel('Day')
plt.ylabel('Sales')
plt.title('Sales over July')
plt.subplot(1, 3, 3)
data[data.month == 'May'].groupby('Day')['Sales'].sum().plot(kind='bar')
plt.xlabel('Day')
plt.ylabel('Sales')
plt.title('Sales over May')
plt.show()
```

<b>TASK 3</b> Scatterplots

Take a look at well known <u>Temperature - Sales relationship</u>, visualized by scatterplot


```python
plt.scatter(data.Temperature, data.Sales, 50, marker = '*', color = '#2Ffd50')
plt.yticks(np.arange(0, 60, 5))
plt.xticks(np.arange(10, 130, 10))
plt.xlabel('Temperature')
plt.ylabel('Sales')
plt.axis('equal')
plt.title('Sales and Temperature relationship')
plt.show()
```
